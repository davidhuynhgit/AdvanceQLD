### 6.2 Unstructured data analysis

#### Data fetching from Guardian API

First, I use Guardian API key to extract article's data from the platform. For the search string, I original think of using Queensland and university or research as they are directly relevant to my concern and questions. According to the vision of Advance Queensland, one of the six priorities is **Research collaboration and commercialisation** so I added those words such as `opportunities,partnership,industry,commercialisation` and limit the fixed timeframe because they narrow down the results to an acceptable number, which is 96, and increase the relevance to the key purpose of the initiative.

```python
#import required libraries
import requests
import json
import re
import time

#load personal API key
with open('guardian_key.txt', 'r') as file:
    key = file.read().strip()

#build a search URL
base_url = 'https://content.guardianapis.com/'
search_string = 'Queensland '
search_string += 'AND (university OR research) '
search_string += 'AND (opportunities OR partnership OR industry OR commercialisation) '
search_string = search_string.replace(' ','%20')
production_office = "aus"
from_date = "2015-09-01"
to_date = "2024-05-01"
full_url = base_url+f"search?&q={search_string}&production-office={production_office}&from-date={from_date}&to-date={to_date}&show-fields=body&api-key={key}"

# get data from server
server_response = requests.get(full_url)
server_data = server_response.json()
resp_data = server_data.get('response','')
if resp_data == '':
    print("ERROR obtaining results:",server_data)
else:
    print("SUCCESS!")
    print(f"{resp_data['total']} results found available in {resp_data['pages']} pages")
    print(f"{resp_data['pageSize']} results per page")
    results = resp_data.get('results',[])
```

To eliminate the impact of duplicate words from the same origin, I use lemmatization algorithm from `nltk` package. I also tried Stemming but that approach may leave my words uninterpretable. Although lemmatization cannot processed the terms thoroughly, it at least gives my nice terms. During that process, only words contain more than 2 characters are kept to build a meaningful pool of words (terms). As we only find the insight as topics, we might also remove the number and digits from articles' content by adding regular expression `<.*?>|[0-9]` from `re` package [example](https://pythonexamples.org/python-re-sub/).

```python
from sklearn.feature_extraction import text
def text_processing(content):
    from nltk.stem import WordNetLemmatizer
    # Initialize Python lemmatizer
    lemmatizer = WordNetLemmatizer()
    processed_string = ''
    for word in content.split():
        if (len(word) > 2) & (word not in list(text.ENGLISH_STOP_WORDS)):           
            processed_string += lemmatizer.lemmatize(word) + ' '
    return processed_string

def articles_from_page_results(page_results):
    articles = {}
    for result in page_results:
        article_date = result['webPublicationDate']
        article_title = result['webTitle']+f" [{article_date}]"
        article_html = result['fields']['body']
        article_text = re.sub(r'<.*?>|[0-9]','',article_html)
        articles[article_title] = text_processing(article_text)
    return articles

def get_all_articles_for_response(response_json,full_url):
    total_pages = response_json['pages']
    total_articles = response_json['total']
    all_articles = {}
    page1_articles = articles_from_page_results(response_json['results'])
    all_articles.update(page1_articles)
    print("Added articles for page: 1")
    
    for page in range(2,total_pages+1):
        page_response = requests.get(full_url+f"&page={page}")
        page_data = page_response.json()['response']
        print("Processing results for page:",page_data['currentPage'])
        page_articles = articles_from_page_results(page_data['results'])
        all_articles.update(page_articles)
        time.sleep(1) # make sure we're not hitting the API to hard
    
    print(f"FINISHED: Fetched {len(all_articles)} articles.")
    return all_articles

my_articles = get_all_articles_for_response(resp_data,full_url)
```

```python
# Load the data - articles from The Guardian
file_path = ""
file_name = "advance_queensland.json"

with open(f"{file_path}{file_name}",'r', encoding='utf-8') as fp:
    articles = json.load(fp)

print(f"Loaded {len(articles)} articles from {file_name}")
```

Then I create a dataframe to store top 10 terms for `tfidf`, `nmf` as I use them for both plotting wordcloud and topic modelling.

```python
# Create a dataframe to hold top terms for each analysis type
terms_df = pd.DataFrame(index=articles.keys(),columns=['tfidf','nmf'])
terms_df
```

#### Term Frequency / Inverse Document Frequency (TF/IDF)


Instead of merely count the term frequency, TF/IDF can help determine how importance a term is in a document relative to a collection of documents. 
The conguration for the Tfidf Vectorizer 0.7, and 0.05 give me the best solution as the dataset is quite big and contains too many words. Instead of putting constants, I set them as percentage to adapt the number of articles. If I increase or decrease, some repetitive insignificant words may mix the result up. Max feature 10000 allows enough room for diversity but ensures the feasibily of computational cost. 

In addition, I put a lot consideration for additional stop words as it is social data, journalists and composers use the names of politicians to express and emphasize their insights very often. Not only some abbreviation but also repetitive terms contribute as the outliers. The specific entities like poliece, parliament, coalition and minister are mentioned many times to enhance the significance of the articles, but I care more about the actions, the specific opinions from public

```python
# Customise stop words
peole_name = ['frydenberg','dutton','anthony','albanese','turnbull','scott','morrison','taylor','wong','cormann','andrews','chalmers','hanson','malcolm','shorten','abbott']
insignificant = ['updated','labor','says','voice','today','tonight','police','parliament','coalition','people','minister']
abbr = ['bst','gmt','am','pm','mdash','bn','nbsp','ve']
my_additional_stop_words = set(peole_name + insignificant + abbr)
my_stop_words = list(text.ENGLISH_STOP_WORDS.union(my_additional_stop_words))

# Set parameters appropriate to your data
tfidf_vectorizer = TfidfVectorizer(max_df=0.7,min_df=.05,max_features=10000,stop_words = my_stop_words)

# Get the document vectors
tfidf_dt_matrix = tfidf_vectorizer.fit_transform(articles.values())

# List of feature names ~ or list of terms|
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a df to combine matrix with feature names
tfidf_df = pd.DataFrame(tfidf_dt_matrix.toarray(), index=articles.keys(), columns=feature_names)
```

Once the matrix of features' weight in documents, by ranking them, we can obtain the top-10 term list for each of document, updated in the global dataframe term_df

```python
for idx in terms_df.index:
    tfidf = dict(tfidf_df.loc[idx].sort_values(ascending=False).head(10))
    #print(counts)
    terms_df.at[idx,'tfidf'] = list(tfidf.keys()) 
terms_df
```

#### Non-negative Matrix Factorisation (NMF)

Despite getting a sense of common topics, if I only rely on terms (which are limited by 10000 by tfidf), I may miss some critical information. As a result, I use NMF algorithm to obtaining topics (a list of terms) from a document-term matrix with 2 factor matrices: document-topic and topic-term. From those two, I can combine and update the term matrix to compare among the methodology. I tried to use LDA but the topics seemed to mixed up and computationally costly. There were also some topic with no documents since the mechanism to classify the documents based on the highest probability. NMF in this case is faster and produce more acceptably relevant topics.

I tried several desired number of topic starting from 10 and reduced it until I obtained the non-overlapping topics. Then we may gain some insight by extracting top-10 terms from each topic. By setting the random state with an interger (I arbitrarily seleted 10), I can exactly reproduce the results.

```python
# Set the number of topics
num_topics = 6

# Create the model
nmf_model = NMF(n_components=num_topics,init='random',beta_loss='frobenius',random_state = 10)

# Fit the model to the data and use it to transform the data
doc_topic_nmf = nmf_model.fit_transform(tfidf_dt_matrix)

# Obtain the topics
topic_term_nmf = nmf_model.components_

# Get the topics and their terms
nmf_topic_dict = {}
for index, topic in enumerate(topic_term_nmf):
    zipped = zip(feature_names, topic)
    top_terms=dict(sorted(zipped, key = lambda t: t[1], reverse=True)[:15])
    #print(top_terms)
    top_terms_list= {key : round(top_terms[key], 4) for key in top_terms.keys()}
    nmf_topic_dict[f"topic_{index}"] = top_terms_list

# Print the topics with their terms    
for k,v in nmf_topic_dict.items():
    print(k)
    print(v)
    print()
```

I searched on Google "topic modelling visualization" and got inspired by the horizontal bar charts. It help me visualise the topics effectively with an eye-catching representation.

Again, I collect selected the most significant topic in each documents and get the corresponding topic-term matrix to extract the term list. Then the term_df is updated with NMF results

```python
for idx,topic in enumerate(doc_topic_nmf):
    topic_num = topic.argmax()
    top_topic = nmf_topic_dict[f"topic_{topic_num}"]
    terms_df['nmf'].iloc[idx] = list(top_topic.keys())
```

To prove the fact that NMF can cluster documents by topics better in my case, I plot the number of documents in each topics. Most of the time I run NMF, every topic contains and remarkable number of documents. I believe this also helpful to show the distribution of results. I think with hundreds of articles, a topic containing only 1 or 2 documents can be interpreted as singularity. Therefore, I did not choose LDA.

```python
topic_doc_distr_dict = {}
for idx,topic in enumerate(doc_topic_nmf):
    topic_num = topic.argmax()
    if topic_num in topic_doc_distr_dict:
        topic_doc_distr_dict[topic_num] += 1
    else:
        topic_doc_distr_dict[topic_num] = 1
topic_doc_distr_df = pd.DataFrame(list(topic_doc_distr_dict.items()), columns=['Topic', 'Number of Articles'])
topic_doc_distr_df_sorted = topic_doc_distr_df.sort_values(by='Number of Articles', ascending=False).reset_index()
topic_doc_distr_df_sorted['Topic'] = topic_doc_distr_df_sorted['Topic']+1
topic_doc_distr_df_sorted['Percentage (%)'] = round(topic_doc_distr_df_sorted['Number of Articles']/len(articles)*100,2)
topic_doc_distr_df_sorted.plot(kind ='bar',x='Topic',y='Percentage (%)',rot=0)
plt.title(label = "Fig 7. Distribution of articles by topics", size = 10,  fontweight="bold")
plt.savefig('graphics/fig7.png',bbox_inches='tight')
plt.show()
```
